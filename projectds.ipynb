```python
import csv

def convert_to_float(value):
    try:
        return float(value)
    except ValueError:
        return None

with open('Sales Data.csv', 'r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)

    # Skip the header if it exists
    next(csv_reader, None)

    for column in csv_reader:
        values = []

        # Iterate through each value in the row
        for value in column:
            converted_value = convert_to_float(value)

            # Check if the conversion was successful (not None)
            if converted_value is not None:
                values.append(converted_value)

        # Check if there are numeric values in the colummn
        if values:
            # Calculate the column average
            column_average = sum(values) / len(values)

            # Print the result
            print( column_average)
```

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df=pd.read_csv('Sales Data.csv')
df.head()
```

```python
a = df['Price Each']
b = df['Sales']
```

```python
t2 = stats.ttest_ind(a,b)
t2
```

```python
#The test statistic is -0.99968598 and the corresponding p-value is 0.3174 . 
#The p-value is less than 0.05, we reject the null hypothesis. 
#Now we have sufficient evidence to say that this data has skewness and kurtosis which is different from a normal distribution
```

```python
from scipy.stats import chi2_contingency #chi-square test

# defining the table
data = [[207, 282, 241], [234, 242, 232]]
stat, p, dof, expected = chi2_contingency(data)

# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```
